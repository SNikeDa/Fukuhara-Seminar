```python
def test_stationarity(data):
    rolling_mean = data.rolling(window=7).mean()
    rolling_std = data.rolling(window=7).std()
    
    fig, ax = plt.subplots(figsize=(12, 8))
    ax.plot(data, label='Original')
    ax.plot(rolling_mean, label='Rolling Mean')
    ax.plot(rolling_std, label='Rolling Std')
    
    ax.legend()
    ax.set_xlabel('Time')
    ax.set_ylabel('Value')
    plt.title('Rolling Mean and Standard Deviation - {}'.format(data.name))
    plt.tight_layout()
    plt.show()
    
    result = adfuller(data)
    print('Augumented Dickey-Fuller Test')
    print('ADF Statistic:', result[0])
    print('p-value:', result[1])
    print('Critical Values:', result[4])
    
def acf_analysis(data):
    auto = sm.tsa.stattools.acf(data, nlags=len(data))

    fig, ax = plt.subplots(figsize=(12, 8))
    plot_acf(data, ax=ax, lags=len(data)-1)
    plt.xlabel('Lag')
    plt.ylabel('Autocorrelation')
    plt.title('Autocorrelation')
    plt.tight_layout()
    plt.show()
    
def pacf_analysis(data):
    partial_auto = sm.tsa.stattools.pacf(data, method='ywm')

    fig, ax = plt.subplots(figsize=(12, 8))
    plot_pacf(data, ax=ax, method='ywm')
    plt.xlabel('Lag')
    plt.ylabel('Partial Autocorrelation')
    plt.title('Partial Autocorrelation')
    plt.tight_layout()
    plt.show()
    
def stl_analysis(data):
    stl = STL(data, period = 12).fit()

    plt.suptitle('Seasonal Trend Decomposition using LOESS')

    plt.subplot(411)
    plt.plot(stl.observed)
    plt.xticks([])
    plt.title('Observed')

    plt.subplot(412)
    plt.plot(stl.trend)
    plt.xticks([])
    plt.title('Trend')

    plt.subplot(413)
    plt.plot(stl.seasonal)
    plt.xticks([])
    plt.title('Seasonal')

    plt.subplot(414)
    plt.plot(stl.resid)
    plt.xticks(rotation=45) 
    plt.title('Residual')

    plt.tight_layout()
    plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.tsa.api import ARIMA
from statsmodels.tools.eval_measures import mse
from statsmodels.tsa.stattools import arma_order_select_ic
from sklearn.metrics import mean_squared_error

def fit_arma(data):
    # Split data into training and test sets
    train_size = int(len(data) * 0.8)
    train, test = data[:train_size], data[train_size:]
    
    # Select the best ARMA order using the arma_order_select_ic function
    order = arma_order_select_ic(train, ic='aic', trend='n')['aic_min_order']
    
    # Convert (p, q) to (p, 0, q)
    ar_order, ma_order = order
    order = (ar_order, 0, ma_order)
    
    # Fit the ARMA model
    model = ARIMA(train, order=order)
    model_fit = model.fit()
    
    # Predict on the test set
    predictions = model_fit.forecast(steps=len(test)).values
    
    # Calculate the Mean Squared Error
    mse = mean_squared_error(test, predictions)
    
    return order, model_fit, mse, test, predictions


# Sample data (replace this with your time series data)
temp_data = data['apur']

# Call the function
order, model_fit, mse_value, test, predictions = fit_arma(temp_data)

# Print the results
print("ARMA Order:", order)
print(model_fit.summary())
print("\nMean Squared Error:", mse_value)

# Plot actual data vs ARMA predictions
plt.figure(figsize=(20, 5))
plt.plot(temp_data, label='Actual Data', color='blue')
plt.plot(test.index, predictions, label='ARMA Predictions', color='red', linestyle='--')
plt.legend()
plt.title('Actual Data vs ARMA Predictions')
plt.grid(True)
plt.show()
```
